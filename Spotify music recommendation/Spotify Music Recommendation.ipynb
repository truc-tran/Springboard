### Explore the data

```python
# Statistic of the data for float and int data types
data.drop(columns=['key', 'mode', 'explicit', 'year']).describe().T.sort_index()
```

```python
# Converting release_date to date_time df 
data['release_date'] =  pd.to_datetime(data['release_date'], format='%Y-%m-%d')
data.head(3)
```

```python
# UNIQUE VALUE
# Check unique value for popularity: missing from 97-99

print('Popularity: \n{}\n'.format(data.sort_values(['popularity']).popularity.unique()))
print('Duration_ms: \n{}\n'.format(data.sort_values(['duration_ms']).duration_ms.unique()))
print('Release date: \n{}\n'.format(data.sort_values(['release_date']).release_date.unique()))
```

```python
# Converting duration_ms to seconds
data['duration'] = data['duration_ms'].apply(lambda x: int(x/1000))
print('Duration (s): \n{}\n'.format(data.sort_values(['duration']).duration.unique()))

# Drop key, mode, explicit, year and duration_ms
df = data.drop(columns=['key', 'mode', 'explicit', 'year', 'duration_ms'])
df.head(3)
```

## Handling NA (missing values) and duplicate

#### Number Of Missing Values By Column

```python
print('Any null values? ', data.isnull().values.any())
print('Checking null values...')
df.isnull().sum()
```

#### Filtering duplicate songs

```python
print('----------Songs with the same name----------')
print(df['name'].value_counts())
print()

print('----------White Christmas----------')
white_christmas = df[df['name'] == 'White Christmas']
white_christmas[['name', 'artists', 'release_date']].sort_values(by=['release_date'])
```

```python
df.shape
```

```python
# Songs with the same name & artirst and release date
print('----------Songs of the same name with artist----------')
print((df['name'] + ', ' + df['artists']).value_counts())
print()

print('----------One Year In A Loop----------')
song = df[df['name'] == 'One Year In A Loop']
song[['name', 'artists', 'release_date']].sort_values(by=['release_date'])
```

#### Remove duplicate songs

```python
# Remove duplicate songs ( same name + same artist)
df = df.drop_duplicates(subset=['name', 'artists'], keep='last')
```

```python
# Songs with the same name & artirst and release date
print('----------Songs of the same name with artist----------')
print((df['name'] + ', ' + df['artists']).value_counts())
print()

print('----------One Year In A Loop----------')
song = df[df['name'] == 'One Year In A Loop']
song[['name', 'artists', 'release_date']].sort_values(by=['release_date'])
```

```python
print('----------Songs with the same name----------')
print(df['name'].value_counts())
print()

print('----------White Christmas----------')
white_christmas = df[df['name'] == 'White Christmas']
white_christmas[['name', 'artists', 'release_date']].sort_values(by=['release_date'])
```
