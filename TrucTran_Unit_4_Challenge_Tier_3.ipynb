**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# return a list of prices ratio for ALL LONDON Bouroughs
def priceRatio(df):
  y2018 = float(df['Average_price'][df['Year'] == 2018])    # Need to convert Series type to Float
  y1998 = float(df['Average_price'][df['Year'] == 1998])
  return [y2018/y1998]           # Must return a list, in order to convert into DataFrame later
```

```python
# Ratio of one district, Ex: Harrow
harrow_ratio = priceRatio(df_groupYear[df_groupYear['London_Borough'] == 'Harrow'])
print(harrow_ratio)
print(type(harrow_ratio))
```

```python
# Check datatype of unique list ==> numpyArray
print(type(df_groupYear['London_Borough'].unique()))
```

```python
# Make a dictionary with Boroughs annd and its ratio
boroughs_price_ratio = {}

for b in df_groupYear['London_Borough'].unique():
  boroughs_price_ratio[b] = priceRatio(df_groupYear[df_groupYear['London_Borough'] == b])

print(boroughs_price_ratio)
```

# Make Final Dataframe with index and price ratio between 2018 and 1998

```python
df_ratios = pd.DataFrame(boroughs_price_ratio)
df_ratios.head()
```

```python
# Formatting and index
df_ratios = df_ratios.T.reset_index()
df_ratios.head(3)
```

```python
# Rename columns
# OR : df_ratios = df_ratios.rename(columns={'index' : 'Borough', 0: 'Ratio'})
df_ratios.rename(columns={'index' : 'Borough', 0: 'Ratio'}, inplace=True)
df_ratios.head(3)
```

# TAKE TOP 15 borough with highest ratio between 2018 and 2019

```python
# using sort_values(by='COL_NAME"), default one has: ascending=True
top15 = df_ratios.sort_values(by='Ratio', ascending=False).head(15)
print(top15)
```

# Bar Graph the top 15

```python
top15_plt = top15.plot(kind='bar', x='Borough', y='Ratio', title='Top 15 London Boroughs Price Ratio between 2018 and 1998')
top15_plt.set_xticklabels(top15.Borough)
top15_plt.set_ylabel('Price ratio')
```
